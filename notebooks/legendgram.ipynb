```python
import geodatasets
import geopandas as gpd
import matplotlib.pyplot as plt
from mapclassify import classify
import mapclassify

gdf = gpd.read_file(geodatasets.get_path("geoda.milwaukee1"))
gdf = gdf.to_crs(gdf.estimate_utm_crs())

c = classify(gdf.HH_INC, scheme='quantiles', k=8)
```

```python
f,ax = plt.subplots(figsize=(8,8))
gdf.plot('HH_INC', scheme='quantiles', k=8, ax=ax)
c.plot_legendgram(ax=ax, loc='upper left', legend_size=("27%", "20%"), inset=True, )
```

```python
f,ax = plt.subplots(1,2, figsize=(10,6))
gdf.plot('HH_INC', scheme='quantiles',ax=ax[0])
c.plot_legendgram(bins=100, inset=False, ax=ax[1], vlines=True, vlinecolor='magenta')
```

```python
f, ax = plt.subplots(figsize=(8, 8))
gdf.plot(
    "HH_INC",
    scheme="quantiles",
    k=8,
    ax=ax,
    legend=True,
    legend_kwds={
        "bbox_to_anchor": (
            1.5,
            1,
        )
    },
)
c.plot_legendgram(ax=ax,
                  loc="right",
                  bbox_to_anchor=(1.5, .5),
                  inset=True,
                  legend_size=(2, 1),
                  )
```

```python
f, ax = plt.subplots(figsize=(8, 8))
gdf.plot(
    "HH_INC",
    scheme="quantiles",
    k=8,
    ax=ax,
    legend=True,
    legend_kwds={
        "bbox_to_anchor": (
            1.5,
            1,
        )
    },
)
c.plot_legendgram(ax=ax,
                  loc="right",
                  bbox_to_anchor=(.4, .0, 1, 1),
                  inset=True,
                  )
```
